```python
#import numpy, matplotlib and iqr from scipy
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import iqr
from scipy.stats import spearmanr
from scipy.stats import kendalltau
```

```python
# read data to a variable
irisdata = open("iris.data","r")

# store lines to a variable for editing purposes
irislines = irisdata.readlines()
```

```python
# delete \n from the last element of each row
for i in range( 0 , len(irislines)):
    irislines[i] = irislines[i][0:len(irislines[i])-1]
```

```python
# separate at ","
irisSep = []

for i in range(0,len(irislines)):
    irisSep.append(irislines[i].split(","))
```

```python
# isolate "Iris-setosa" rows
irisSep_setosa = []

for i in range(0,len(irisSep)):
    if (irisSep[i][len(irisSep[i])-1] == "Iris-setosa"):
        irisSep_setosa.append(irisSep[i][0:4])
```

```python
# Convert values to float
irisSep_setosa = np.array(irisSep_setosa, dtype = float)
```

```python
# mean of each attribute
print("Mean of each attribute:")
print("  sepal length in cm: "+str(np.mean(irisSep_setosa, axis = 0)[0]))
print("  sepal width in cm: "+str(np.mean(irisSep_setosa, axis = 0)[1]))
print("  petal length in cm: "+str(np.mean(irisSep_setosa, axis = 0)[2]))
print("  petal width in cm: "+str(np.mean(irisSep_setosa, axis = 0)[3]))
```

```python
# median of each attribute
print("Median of each attribute:")
print("  sepal length in cm: "+str(np.median(irisSep_setosa, axis = 0)[0]))
print("  sepal width in cm: "+str(np.median(irisSep_setosa, axis = 0)[1]))
print("  petal length in cm: "+str(np.median(irisSep_setosa, axis = 0)[2]))
print("  petal width in cm: "+str(np.median(irisSep_setosa, axis = 0)[3]))
```

```python
# standard deviation of each attribute
print("Standard deviation of each attribute:")
print("  sepal length in cm: "+str(np.std(irisSep_setosa, axis = 0)[0]))
print("  sepal width in cm: "+str(np.std(irisSep_setosa, axis = 0)[1]))
print("  petal length in cm: "+str(np.std(irisSep_setosa, axis = 0)[2]))
print("  petal width in cm: "+str(np.std(irisSep_setosa, axis = 0)[3]))
```

```python
# isolate "Iris-virginica" rows
irisSep_virginica = []

for i in range(0,len(irisSep)):
    if (irisSep[i][len(irisSep[i])-1] == "Iris-virginica"):
        irisSep_virginica.append(irisSep[i][0:4])
        
# Convert values to float
irisSep_virginica = np.array(irisSep_virginica, dtype = float)

# isolate "Iris-versicolor" rows      
irisSep_versicolor = []

for i in range(0,len(irisSep)):
    if (irisSep[i][len(irisSep[i])-1] == "Iris-versicolor"):
        irisSep_versicolor.append(irisSep[i][0:4])
        
# Convert values to float        
irisSep_versicolor = np.array(irisSep_versicolor, dtype = float)
```

```python
# plots the histogram of dataset x using the four binsizes defined in the lecture material
# histogram x axis labeled with xlabel
def quad_plot ( x , xlabel ):
    # define frequently used values as variables for convenience
    dataset1 = np.transpose(x)[0]
    mini = np.min(dataset1)
    maxi = np.max(dataset1)
    size = len(dataset1)

    # Sturges' rule
    k_sturges = int(np.log2(size) + 1)+1
    plt.subplot(2, 2, 1)
    plt.hist(dataset1, bins=np.linspace(mini,maxi,num=k_sturges),edgecolor="black")
    plt.title("Strurges' rule\n k = "+str(k_sturges))
    plt.ylabel("Counts")
    plt.xlabel(xlabel)

    # Scott's rule
    k_scott = int((maxi - mini)/((3.5*np.std(dataset1))/(size**(1/3))))+1
    plt.subplot(2, 2, 2)
    plt.hist(dataset1, bins=np.linspace(mini,maxi,num=k_scott),edgecolor="black")
    plt.title("Scott's rule\n k = "+str(k_scott))
    plt.ylabel("Counts")
    plt.xlabel(xlabel)

    # Square-root choice
    k_sqrt = int(np.sqrt(size))+1
    plt.subplot(2, 2, 3)
    plt.hist(dataset1, bins=np.linspace(mini,maxi,num=k_sqrt),edgecolor="black")
    plt.title("Square-root choice\n k = "+str(k_sqrt))
    plt.ylabel("Counts")
    plt.xlabel(xlabel)

    # Freedman-Diaconis' choice
    k_FD = int(((size**(1/3))*(maxi - mini))/(2*iqr(dataset1, rng = (25,75))))+1
    plt.subplot(2, 2, 4)
    plt.hist(dataset1, bins=np.linspace(mini,maxi,num=k_FD),edgecolor="black")
    plt.title("Freedman-Diaconis' choice\n k = "+str(k_FD))
    plt.ylabel("Counts")
    plt.xlabel(xlabel)

    plt.tight_layout()
    plt.show()
```

```python
print("Histograms for Iris Setosa")
quad_plot( irisSep_setosa, "Sepal length in cm" )
```

Scott's rule clearly results in too few bins. Strurges' rule gives a bit more shape for the distribution, and makes it look like approximately symmetric w.r.t. middle value. However, with the square-root choice and Freedman-Diaconis' choice, we see that the distribution is actually leaning to the right of the middle value.

```python
# isolate number rows
irisSep_all = []

for i in range(0,len(irisSep)-1):
        irisSep_all.append(irisSep[i][0:4])

irisSep_all = np.array(irisSep_all, dtype = float)
```

```python
# Pearson correlation
print("Pearson correlation matrix\n")
print(np.corrcoef(np.transpose(irisSep_all)))
```

Pearson correlation, from Wikipedia: The correlation coefficient ranges from 1 to 1. A value of 1 implies that a linear equation describes the relationship between X and Y perfectly, with all data points lying on a line for which Y increases as X increases. A value of 1 implies that all data points lie on a line for which Y decreases as X increases. A value of 0 implies that there is no linear correlation between the variables.

From the matrix we can see that the 3rd and 4th attribute correlate heavily with each other. Also, the second attribute is anti-correlated with all other attributes.

```python
# Spearman Rho
print("Spearman Rho\n")
print(spearmanr(irisSep_all)[0])
```

Sperman Rho from Wikipedia: The sign of the Spearman correlation indicates the direction of association between X (the independent variable) and Y (the dependent variable). If Y tends to increase when X increases, the Spearman correlation coefficient is positive. If Y tends to decrease when X increases, the Spearman correlation coefficient is negative. A Spearman correlation of zero indicates that there is no tendency for Y to either increase or decrease when X increases.

The Spearman Rho correlation coefficient indicates, how well we can approximate the given data using a monotonic function. We can see very similar relationship between different attributes as before with Pearson correlation.

```python
# Kendall's Tau
print("Kendall Tau\n")

kennmatrix = np.zeros((4,4),dtype=float)

for i in range(0,len(kennmatrix)):
    for j in range(0,len(kennmatrix[i])):
        kennmatrix[i][j] = kendalltau(np.transpose(irisSep_all)[i],np.transpose(irisSep_all)[j])[0]

print(kennmatrix)
```

Kendall's Tau from Wikipedia: Intuitively, the Kendall correlation between two variables will be high when observations have a similar (or identical for a correlation of 1) rank (i.e. relative position label of the observations within the variable: 1st, 2nd, 3rd, etc.) between the two variables, and low when observations have a dissimilar (or fully different for a correlation of -1) rank between the two variables.

This seems that the Kendall Tau is a sort of a sum of the pearson correlations between pairs of values in the scatter plot. Again the signs of the correlations agree with the previous two.

```python
# Make a copy of the original array for all flower data
irisSep_norm_t = np.zeros((len(irisSep_all.transpose()),len(irisSep_all.transpose()[0])))

for k in range(0,len(irisSep_all.transpose())):
    for h in range(0,len(irisSep_all.transpose()[k])):
        irisSep_norm_t[k][h] = irisSep_all.transpose()[k][h]

# Center the data
for i in range(0,len(irisSep_norm_t)):
    for j in range(0,len(irisSep_norm_t[i])):
        irisSep_norm_t[i][j] = (irisSep_norm_t[i][j] - np.transpose(irisSep_all)[i].mean())/(np.std(irisSep_all[i]))

# Convert to numpy array
irisSep_norm_t = np.array(irisSep_norm_t , dtype = float)
```

```python
# Correlation matrix
C = (1/((np.transpose(irisSep_all).shape[0])-1))*np.matmul(irisSep_norm_t,irisSep_norm_t.transpose())
```

```python
# List eigenvalues and vectors
np.linalg.eigh(C)
```

```python
eig1 = np.linalg.eigh(C)[0][3]
eig2 = np.linalg.eigh(C)[0][2]
M = np.linalg.eigh(C)[1][0:4,2:4]
```

```python
# Projection matrix with vectors corresponding to the largest
# two eigenvalues
M
```

```python
# Plot PCA results
plt.figure(figsize=(8,8))
plt.scatter(np.matmul(M.transpose(),irisSep_norm_t)[0][0:50],np.matmul(M.transpose(),irisSep_norm_t)[1][0:50], color="blue", label="Setosa")
plt.scatter(np.matmul(M.transpose(),irisSep_norm_t)[0][50:100],np.matmul(M.transpose(),irisSep_norm_t)[1][50:100], color="green", label="Versicolor")
plt.scatter(np.matmul(M.transpose(),irisSep_norm_t)[0][100:150],np.matmul(M.transpose(),irisSep_norm_t)[1][100:150], color="orange", label="Virginica")
plt.legend(loc='right')
plt.grid(True,alpha=0.2)
plt.show()
```
